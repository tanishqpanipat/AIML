```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
X = iris.data
```

```python
numSamples, numFeatures = iris.data.shape
print (numSamples)
print (numFeatures)
print (iris.target_names)
```

```python
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)
```

```python
X= iris.data
pca = PCA(n_components=2, whiten=True).fit(X) # Set the number of components to reduce to
X_pca = pca.fit_transform(X_standardized)  # Fit and transform the data
```

```python
print(pca.components_)
```

```python
print(pca.explained_variance_ratio_)
print(sum(pca.explained_variance_ratio_))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=iris.target, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.colorbar(label='Target Class')
plt.show()
```

```python
explained_variance = pca.explained_variance_ratio_
print(f"Explained Variance of PC1: {explained_variance[0]:.2f}")
print(f"Explained Variance of PC2: {explained_variance[1]:.2f}")
```

Explained variance in a dataset is a measure of how much of the total variance or variability in the data can be accounted for or "explained" by a specific factor or component. In the context of dimensionality reduction techniques like Principal Component Analysis (PCA), explained variance is often used to evaluate how well the principal components capture the important information in the data.
