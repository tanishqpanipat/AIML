```python
import pandas as pd
import numpy as np
import seaborn as sns
url = 'https://raw.githubusercontent.com/tanishqpanipat/AIML/main/bmi_train.csv'
df = pd.read_csv(url)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['Weight']] = scaler.fit_transform(df[['Weight']])
```

```python
#checking if threre are duplicates or not
a = df.duplicated().any()
#print(a)
if(a):
    print('There are duplicates')
else:
    print('No duplicates')
```

```python
#Detecting outliers
q1=df.quantile(0.25)
q3=df.quantile(0.75)
iqr=q3-q1
outliers=(df>q3+3*iqr) | (df<q1-3*iqr)
k=outliers.sum()
print(k)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table between 'Height' and 'Weight'
contingency_table = pd.crosstab(df['Height'], df['Weight'])

# Perform the chi-square test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")

# Interpret the results (e.g., check if p < 0.05 for significance)
if p < 0.05:
    print("There is a significant association between Height and Weight.")
else:
    print("There is no significant association between Height and Weight.")
```
