```python
import pandas as pd
import numpy as np
url = 'https://raw.githubusercontent.com/tanishqpanipat/AIML/main/movies.csv'
df = pd.read_csv(url)
```

```python
df
```

```python
# Separate the target variable (assuming 'target' is your target column)
data= df.copy()

# Calculate Pearson's correlation coefficients
correlation_matrix = np.abs(data.corr())

# Set a correlation threshold
correlation_threshold = 0.3

# Select features with correlation above the threshold to 'injured'
selected_features = correlation_matrix['runtime'][correlation_matrix['runtime'] > correlation_threshold]

# Filter the dataset to keep only the selected features
X_filtered = data[selected_features.index]

# Print the selected features
print("Selected Features:")
print(selected_features.index)
```

Pearson's feature selection method, also known as Pearson correlation coefficient or Pearson's correlation, is a statistical technique used to measure the linear relationship between two continuous variables. It quantifies the strength and direction of the linear association between two variables. This method is often used in feature selection for machine learning and data analysis to identify which features (variables) are most strongly correlated with a target variable or with each other.
